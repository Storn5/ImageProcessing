```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
```

## Sliding the image histogram to the right
When sliding the histogram to the right, we add a value to every pixel's intensity in the image, making the image brighter. If the value exceeds 255, it is reset to 255.


```python
def slide_hist_right(img, value):
    # Convert type from uint8 to uint16 to be able to go over 255
    res_img = np.asarray(img, dtype=np.uint16)
    # Add value to each pixel, but don't go over 255
    res_img = np.minimum(255, res_img + value)
    # Convert back to uint8
    return np.asarray(res_img, dtype=np.uint8)
```

## Reading images

```python
img_low_detail = image.imread("low_detail.jfif")
img_high_detail = image.imread("high_detail.jpg")
img_low_contrast = image.imread("low_contrast.jpg")
img_high_contrast = image.imread("high_contrast.jpg")
```

```python
def show_img(img, title):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot()
    imgplot = plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    ax.set_title(title)
    plt.show()
```

### Sliding images to the right by 50

```python
show_img(img_low_detail, 'Original')
show_img(slide_hist_right(img_low_detail, 50), 'Slide right by 50')
```

```python
show_img(img_high_detail, 'Original')
show_img(slide_hist_right(img_high_detail, 50), 'Slide right by 50')
```

```python
show_img(img_low_contrast, 'Original')
show_img(slide_hist_right(img_low_contrast, 50), 'Slide right by 50')
```

```python
show_img(img_high_contrast, 'Original')
show_img(slide_hist_right(img_high_contrast, 50), 'Slide right by 50')
```
